```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Blastp results
# All proteins
blastp_results = pd.read_csv("../03_blastp/blastp_prodigal_Fagoma_Raton_5498_outfmt6_cov-id_values.tsv", sep="\t")

sns.barplot(data=blastp_results, x="cov-id", y="matches")
plt.title("Ig-like proteins, blastp")
plt.xlabel("")
plt.xticks(rotation=90)
plt.ylabel("Number of matches to Ig-like proteins")
plt.savefig("./blastp_results.png")
plt.show()

# Considering contigs only
blastp_results = pd.read_csv("../03_blastp/blastp_prodigal_Fagoma_Raton_5498_outfmt6_cov-id_contigs.tsv", sep="\t")

sns.barplot(data=blastp_results, x="cov-id", y="contigs")
plt.title("Contigs with Ig-like proteins, blastp")
plt.xlabel("")
plt.xticks(rotation=90)
plt.ylabel("Number of contigs with Ig-like proteins")
plt.savefig("./blastp_results_contigs.png")
plt.show()

```
